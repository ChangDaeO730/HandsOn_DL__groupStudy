### tensor to numpy, numpy to tensor
* numpy() 함수로 텐서를 ndarray로 변환
* 이후 배열에 대해 다양한 함수를 적용할 수 있다.
* tf.convert_to_tensor() 함수를 이용하면 array를 다시 tensor로 변환

```python
import tensorflow as tf
import numpy as np

a = tf.constant(3)
b = tf.constant(2)
c = tf.add(a,b).numpy() 

c_square = np.square(c, dtype = np.float32)
c_tensor = tf.convert_to_tensor(c_square) 

print(c, c_square, c_tensor, sep = "\n")
```

### @tf.function
자동으로 그래프를 생성해주는 기능 
* if -> tf.cond
* for/while -> tf.while_loop
* for _ in dataset -> dataset.reduce
> 파이썬으로 구성된 함수를 고효율의 텐서플로 그래프 형태로 다루고 싶을 때 사용

```python
import tensorflow as tf
@tf.function
def square_pos(x):
    if x > 0:
        x = x * x
    else:
        x = x*-1
    return x

print(square_pos(tf.constant(2)))
print(square_pos)
```

### 3. gradient descent
* 학습률은 학습속도와 성능에 큰 영향을 미침. 
* 초기에 좋은 학습률을 선택하는 것이 중요함. 경험적으로 0.001을 많이 사용
* 다양한 방식의 학습률 scheduling을 선택할 수 있음. 늘렸다 줄였다 해보는게 좋음 (1-cycle scheduling)
* 가중치 초기화방식도 중요. Glorot, he, Lecun 등

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")

lr_list = [0.001, 0.1, 0.5, 0.9]

def get_derivative(lr):
    w_old = 2                            # initial weight
    weights = [w_old]
    
    y = [w_old ** 2]                     # loss function : y = w^2
    
    for i in range(1, 10):
        dev_value = w_old * 2            # derivative
        
        w_new = w_old - lr * dev_value   # update
        w_old = w_new
        
        weights.append(w_old)            # weight logs
        y.append(w_old ** 2)             # loss logs
        
    return weights, y


# visualize
x = np.linspace(-2, 2, 50)
x_square = [i ** 2 for i in x]

fig = plt.figure(figsize = (12, 7))

for i, lr in enumerate(lr_list):
    weights, y = get_derivative(lr)
    ax = fig.add_subplot(2,2, i+1)
    ax.scatter(weights, y, color = "red")
    ax.plot(x, x_square)
    ax.title.set_text('lr = ' + str(lr))

plt.show()
```

```python
tf.random.set_seed(226) # for reproductivity

# sample data
x = np.array([[0,0], [1,0], [0,1], [1,1]])
y = np.array([[0], [1], [1], [0]])

import pandas as pd

opt_list = [RMSprop(lr = 0.1), RMSprop(lr = 0.000001)]
fig = plt.figure(figsize = (20, 8))

for i,opt in enumerate(opt_list):
    model = Sequential()
    model.add(Dense(32, input_shape = (2,), activation = "relu"))
    model.add(Dense(1, activation = "sigmoid"))
    
    model.compile(optimizer = opt,
                  loss = mse,
                  metrics = ['acc'])
    history = model.fit(x, y, epochs = 100, verbose = 0)
    
    ax = fig.add_subplot(1,2, i+1)
    ax.plot(pd.DataFrame(history.history))
```
